1. Uma bola é lançada ao ar. Suponha que sua altura $h(t)$, em metros, e $t$ segundos após o lançamento, seja:

$$h(t) = −t^2 + 4t$$
   
   a) Faça o gráfico da altura em função do tempo. Não se esqueça de indicar o significado de cada eixo, colocando a unidade da altura e do tempo.  
   b) Calcule a altura máxima atingida pela bola.

Sugestão: fazer o gráfico com o tempo entre 0 $s$ e 5 $s$ com incrementos de 0.1 $s$.

```python
import numpy as np
import matplotlib.pyplot as plt

print("a)")

t = np.arange(0, 5, 0.1)
h = -t**2 + 4*t

#Gráfico
plt.plot(t, h, linestyle = '--', color = 'purple')
plt.xlabel('Tempo (s)')
plt.ylabel('Altura (m)')
plt.title('Gráfico da Altura em função do Tempo')
plt.grid(True)
plt.show()

#Tempo para atingir a altura máxima

print("b)")

tempo_max = 2
altura_max = -tempo_max**2 + 4*tempo_max
print("A altura máxima atingida:", altura_max,"m")
```

2. A tensão elétrica medida em um circuito, em volts, tem a seguinte expressão no instante $t$, em segundos:

$$v(t) = e^{-0.5t}\cos\left(2\pi0.8t\right)$$

   a) Faça o gráfico da tensão em função do tempo.
    
Sugestão: fazer o gráfico com tempo entre 0 e 10 $s$ com incremento de 0.05 $s$.

```python
import numpy as np
import matplotlib.pyplot as plt

print("a)")

t = np.arange(0, 10, 0.05)
v = np.exp(-0.5 * t) * np.cos(2 * np.pi * 0.8 * t) #Fórmula

#Gráfico
plt.figure()
plt.plot(t, v, color = 'green')
plt.xlabel('Tempo (s)')
plt.ylabel('Tensão (v)')
plt.title('Gráfico da Tensão Elétrica em um Circuito em função do Tempo')
plt.grid(True)
plt.show()
```

3. Faça um gráfico da equivalência da temperatura em Fahrenheits em função da temperatura em Celsius:

$$F = \frac{9}{5}C + 32 $$

   a) Faça o gráfico tomando 100 pontos entre -20 $^\circ$C e 100 $^\circ$C.

```python
import numpy as np
import matplotlib.pyplot as plt

print("a)")

c = np.arange(-20, 100, 1.2)
f = (c / 5) * 9 + 32

#Gráfico
plt.figure()
plt.plot(c, f, color = 'orange', linestyle = '--')
plt.xlabel('Celsius (ºC)')
plt.ylabel('Fahrenheits (ºF)')
plt.title('Gráfico da Equivalência da temperatura em Fahrenheits em função da temperatura em Celsius')
plt.grid(True)
plt.show()
```

4. Entre os anos de 1994 e 2022, a população de uma cidade cresce aproximadamente de acordo com a seguinte função:

$$P(a) = 800000e^{\frac{a - 1994}{40}}$$

em que $a$ é o ano e $P$ é o número de habitantes.

   a) Faça o gráfico do número de habitantes entre 1994 e 2022.

```python
import numpy as np
import matplotlib.pyplot as plt

print("a)")

a = np.arange(1994, 2022, 1)
p = 800000*np.exp((a - 1994) / 40)

#Gráfico
plt.figure()
plt.plot(a, p, marker = 'd', color = 'turquoise')
plt.xlabel('Anos (a)')
plt.ylabel('População (p)')
plt.title('Gráfico do Número de Habitantes entre 1994 e 2022')
plt.grid(True)
plt.show()
```
