```python
total_por_língua = tmdb["original_language"].value_counts()
total_do_geral = total_por_língua.sum()
total_de_ingles = total_por_língua.loc["en"]
total_do_resto = total_do_geral - total_de_ingles
print(total_do_geral, total_de_ingles, total_do_resto)
```

```python
dados = {
    "língua" : ["inglês", "outros"],
    "total" : [total_de_ingles,total_do_resto]
}
dados = pd.DataFrame(dados)
dados
```

```python
sns.barplot(data=dados, x="língua", y="total")
```

```python
dados.plot(kind="pie", y="total", labels=dados["língua"])
```

```python
total_de_outros_filmes_por_lingua = tmdb.query("original_language != 'en'")["original_language"].value_counts()
total_de_outros_filmes_por_lingua.head()
```

```python
sns.countplot(data=tmdb.query("original_language != 'en'"),
              order=total_de_outros_filmes_por_lingua.index,
              hue="original_language",
              x="original_language")
```

```python
sns.color_palette("mako")
```

```python
total_de_outros_filmes_por_lingua = tmdb.query("original_language != 'en'")["original_language"].value_counts(normalize=True)
total_de_outros_filmes_por_lingua.head()
```

```python
plt.figure(figsize=(16, 8))
sns.countplot(data=tmdb.query("original_language != 'en'"),
              order=total_de_outros_filmes_por_lingua.index,
              palette="mako",
              hue="original_language",
              hue_order=total_de_outros_filmes_por_lingua.index,
              stat="percent",
              x="original_language")
plt.title("Distribuição dos idiomas original nos filmes - exceto inglês")
plt.show()
```

```python
import numpy as np

filme1 = [2.5]*10 + [3.5] * 10
filme2 = [5] * 10 + [1] * 10

media_filme1 = np.mean(filme1)
mediana_filme1 = np.median(filme1)

mediana_filme2 = np.mean(filme2)
mediana_filme2 = np.median(filme2)

print("\nFilme1:")
print("Média:", media_filme1)
print("Mediana:", mediana_filme1)

print("\nFilme2:")
print("Média:", mediana_filme2)
print("Mediana:", mediana_filme2)
```
